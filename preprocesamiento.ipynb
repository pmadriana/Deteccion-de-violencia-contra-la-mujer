```python
import re
import unicodedata
nombre_archivo = "positivo.csv"
lista_data = []
with open(nombre_archivo, "r", encoding="utf8", errors='ignore') as archivo:
    next(archivo)
    for linea in archivo:
        linea = linea.strip('\n')
        linea = unicodedata.normalize("NFKD", linea).encode("ascii","ignore").decode("ascii")
        linea = re.sub("@([a-zA-Z0-9_]{1,50})","",linea)
        linea=linea.split(';')
        lista_data.append(linea)
        print(linea)
```

```python
columnas = ["id","tweet", "text_processed","class"]
tweets_df = pd.DataFrame(lista_data, columns = columnas)
tweets_df.dropna()
tweets_df = tweets_df.drop_duplicates()
tweets_df.head(10)
```

```python
tweets_df.count()
```

```python
import nltk
nltk.download('popular')
```

```python
from nltk.tokenize import word_tokenize
# Lemmatization 
from nltk.stem import WordNetLemmatizer
# Load the regular expression library
import re

# Eliminar RT @  enlaces
tweets_df['text_processed'] = tweets_df['tweet'].map(lambda x: re.sub('RT|cc', '', x))
tweets_df['text_processed'] = tweets_df['text_processed'].map(lambda x: re.sub('@([a-zA-Z0-9_]{1,50})',"",x))
#tweets_df['text_processed'] = tweets_df['text_processed'].map(lambda x: re.sub(r'http\S+'', '', x))
#Eliminar signos de puntuacion
tweets_df['text_processed'] = tweets_df['text_processed'].map(lambda x: re.sub('[,\"|\'.!?]', '', x))
#quitar enlaces
tweets_df['text_processed'] = tweets_df['text_processed'].map(lambda x: re.sub(r"http\S+", '', x))

# Pasar a minusculas
tweets_df['text_processed'] = tweets_df['text_processed'].map(lambda x: x.lower())
# quitar todos los tokens no alfabeticos
tweets_df['text_processed'] = tweets_df['text_processed'].map(lambda x: ' '.join([word for word in word_tokenize(x) if word.isalpha()]))

# eliminar stopwords
sr = nltk.corpus.stopwords.words('spanish')
tweets_df['text_processed'] = tweets_df['text_processed'].map(lambda x: ' '.join([word for word in word_tokenize(x) if not word in sr]))


tweets_df.head(50)
```

```python
tweets_df.to_csv('positivo_limpio.csv', index=False, header=True, encoding="utf-8")
tweets_df.count()
```
